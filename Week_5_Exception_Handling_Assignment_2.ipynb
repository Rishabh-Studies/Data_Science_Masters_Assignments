# Q1. Explain why we have to use the Exception class while creating a Custom Exception.

All Exceptions in python are derived from the base class: Exception. Thus, for defining a custom exception, it's class must be derived from the Exception base class as well.

# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

ArithmeticError Class is called when an error occurs during numeric calculations. There are three errors that are defined in this class:
1. ZeroDivisionError
2. OverflowError
3. FloatingPointError

```python
try:
    10/0
except ZeroDivisionError as e:
    print(e)
```

```python
import math
try:
    print(math.exp(1000))
except OverflowError as e:
    print(e)
```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError exception class acts as a base class for KeyError and IndexError exception classes. This exception class is used when a key or an index is not found in a dictionary or sequence, respectively.

```python
#example of IndexError
l=[1,2,3]
try:
    print(l[3])
except IndexError as e:
    print(e)
```

```python
#example of KeyError
my_dict={1:'a',2:'b'}
try:
    print(my_dict[3])
except KeyError:
    print("KeyError")
```

# Q5. Explain ImportError. What is ModuleNotFoundError?

An ImportError is detected when Python has problems with a successful module import. Usually this problem is caused by the incorrect path and is usually displayed with the message that there is “No module named (…)” or “cannot import name (…)”.

ModuleNotFoundError is a subclass of ImportError exception class. It is raised when a module is not found.

# Q6. List down some best practices for exception handling in python.

1) Avoid using 'Exception' as the class for handling exceptions.
2) Use specific exceptions that are expected to occur.
3) Use the generalized 'Exception' class in the end after mentioning the except block for specific exceptions.
4) Log the exceptions that occur before fixing them.
5) Finally block must recover the resources used.
6) Use understandable exception messages.
